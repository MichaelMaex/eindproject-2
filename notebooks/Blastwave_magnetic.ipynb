```python
import numpy as np
import sympy as sp
from IPython.display import display
import matplotlib as mpl
mpl.rcParams["figure.dpi"] = 240
import matplotlib.pyplot as plt
sp.init_printing()
import pandas as pd
import pyPLUTO as pp
from matplotlib import colors, cm

L0 = 14959789200000.0 # 1 code unit of length in centimeter
v0 = 1e5 # 1 code unit of velocity in cm/s
t0 = L0/v0 # 1 code unit of time in seconds
rho0 = 1
```

# Comparing magnetic field strengths

```python
j = 100
betas = ["0.1", "0.3", "0.5", "1", "5", "10"]
fig, ax, = plt.subplots(2,3, figsize = (15,8))

images = [pp.pload(j, "../pluto-definitions/blastwave_MHD/beta{}-low/".format(beta)) for beta in betas]
im = [0]*6

vmin = min(image.prs.min() for image in images)
vmax = max(image.prs.max() for image in images)
norm = colors.Normalize(vmin=vmin, vmax=vmax)

for i in range(6):
    axis = ax[i//3][i%3]
    beta = betas[i]
    frame = images[i]
    im[i] = axis.imshow(frame.prs, norm=norm, cmap=cm.plasma)
    axis.set_title("beta = {}".format(beta))
fig.tight_layout()
fig.colorbar(im[0], ax  = ax)
fig.savefig("../report/figures/influence_beta.pdf")
```

# Analysing Alven waves

```python
j = 200
T = list(map(lambda t: int(j*t/5), range(6)))
fig, ax, = plt.subplots(2,3, figsize = (15,8))

images = [pp.pload(t, "../pluto-definitions/blastwave_MHD/beta0.1-low/") for t in T]
im = [0]*6

vmin = min(image.prs.min() for image in images)
vmax = max(image.prs.max() for image in images)
norm = colors.Normalize(vmin=7.5, vmax=10)

for i in range(6):
    axis = ax[i//3][i%3]
    t = T[i]
    frame = images[i]
    im[i] = axis.imshow(frame.prs, norm=norm, cmap=cm.plasma)
    axis.set_title("t = {:.2e} (s)".format(frame.SimTime*t0))
fig.tight_layout()
fig.colorbar(im[0], ax  = ax)
fig.savefig("../report/figures/alven_wave.pdf")
```

```python
def radius(img):
    domainsize = len(img)
    background_prs = img[0,0]
    wave_position = np.zeros(domainsize)
    center_line = img[:,domainsize//2]
    j = np.argmax(-center_line)
    return domainsize//2 - j
```

```python
R = np.zeros(200)
t = np.zeros(200)
B0 =  pp.pload(0, "../pluto-definitions/blastwave-MHD/beta0.1-low/").Bx1[0,0]
for i in range(0,200):
    frame = pp.pload(i+1, "../pluto-definitions/blastwave-MHD/beta0.1-low/")
    R[i] = radius(frame.vx1)*frame.dx1[0]
    t[i] = frame.SimTime
```
